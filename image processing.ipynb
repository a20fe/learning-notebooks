```python
import cv2
from matplotlib import pyplot as plt
```

```python
image=cv2.imread('image.jpg')
gray_image =cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.title('asmaa')
plt.axis('off')
plt.show()
           
```
